```python
import numpy as np
```

```python
arr=np.array([1,2,3,6,4,5])
new=arr[::-1]
print(new)
```

```python
array1=np.array([[1,2,3], [2,4,5], [1,2,3]])
print(array1.flatten())
```

```python
arr1=np.array([[1,2], [3,4]])
arr2=np.array([[1,2], [3,4]])
np.array_equal(arr1, arr2)
np.array_equiv(arr1, arr2)
```

```python
gfg=np.matrix('[4,1,9;12,3,1;4,5,6]')
print(gfg.sum())
print(np.sum(gfg, axis=0))
print(np.sum(gfg, axis=1))
```

```python
n_array=np.array([[55,25,15], [30, 44, 2], [11, 45, 77]])
print(np.trace(n_array))
np.linalg.eig(n_array)
```

```python
np.linalg.inv(n_array)
```

```python
np.linalg.det(n_array)
```

```python
p=np.array([[1,2], [2, 3]])
q=np.array([[4,5], [6, 7]])
print(np.dot(p, q))
p2=np.array([[1,2], [2,3], [4,5]])
q2=np.array([[4,5,1], [6,7,2]])
np.cov(p.T, q.T)
```

```python
x=np.array([[2, 3, 4], [3, 2, 9]])
y=np.array([[1,5,0], [5, 10, 13]])
print("Inner:", np.inner(x, y))
print("Outer:", np.outer(x, y))
print("Cartesian:", np.cross(x, y))
```

```python
array=np.array([[1, -2, 3], [-4, 5, -6]])
print("Absolute: ", np.abs(array))
f=array.flatten();
print(np.percentile(f, 25))
print(np.percentile(f, 50))
print(np.percentile(f, 75))
print(f.mean())
print(np.median(f))
print(np.std(f))
```

```python
a=np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
print(np.floor(a))
print(np.ceil(a))
print(np.trunc(a))
print(np.round(a))
```

```python
array=np.array([10, 52, 62, 16, 16, 54, 453])
array.sort()
print(array)
np.sort(array)[:4]
np.sort(array)[-5:]
```

```python
array4=np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
print(array4[array4==np.floor(array4)])
print(array4[array4!=np.ceil(array4)])
```
